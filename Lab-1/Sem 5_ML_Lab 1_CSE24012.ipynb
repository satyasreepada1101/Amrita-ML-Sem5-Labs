```python
# Q1
def count_vowels(text):
    vowels = "aeiou"
    vc = 0
    cc = 0

    for ch in text:
        if ch.isalpha():
            if ch.lower() in vowels:
                vc += 1
            else:
                cc += 1

    return vc, cc


s = input("enter string: ")

v, c = count_vowels(s)

print("num of vowels:", v)
print("num of consonants:", c)
```

```python
# Q2
def matrix_multiply(a, b):

    r1 = len(a)
    c1 = len(a[0])

    r2 = len(b)
    c2 = len(b[0])

    if c1 != r2:
        return "matrices cant be multiplied"

    result = []

    for i in range(r1):
        row = []
        for j in range(c2):
            total = 0
            for k in range(c1):
                total += a[i][k] * b[k][j]
            row.append(total)
        result.append(row)

    return result


a = [
    [1, 2, 3],
    [4, 5, 6]
]

b = [
    [7, 8],
    [9, 10],
    [11, 12]
]

ans = matrix_multiply(a, b)

for i in ans:
  print(i)
```

```python
# Q3
def count_common(list1, list2):

    count = 0

    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                count += 1
                break

    return count


list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common = count_common(list1, list2)

print("no of common eles:", common)
```

```python
# Q4
def transpose_matrix(mat):

    r = len(mat)
    c = len(mat[0])

    trans = []

    for j in range(c):
        row = []
        for i in range(r):
            row.append(mat[i][j])
        trans.append(row)

    return trans


mat = [
    [1, 2, 3],
    [4, 5, 6]
]

ans = transpose_matrix(mat)

print("transpose matrix:")

for i in ans:
    print(i)
```

```python
# Q5
import random
import statistics

def stats(nums):

    mean = statistics.mean(nums)
    median = statistics.median(nums)
    mode = statistics.mode(nums)

    return mean, median, mode

nums = []

for i in range(100):
    nums.append(random.randint(100, 150))

mean, median, mode = stats(nums)

print("random numbers:")
print(nums)

print()
print("mean:", mean)
print("median:", median)
print("mode:", mode)
```
